```python
#import necessary modules for csv files, linear regressions, bayes, and basic plotting
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from matplotlib import pyplot as plt
```

```python
#set up player df for regressions
player_stats=pd.read_csv('players_all_in_one.csv')
player_columns=player_stats.columns
for i in player_columns:
    player_stats[i]=player_stats[i].fillna(int(player_stats[i].mean()))
```

```python
#players stats linear regression vs TS%: 1000 trials
for i in player_columns:
    results=[]
    for a in list(range(1000)):
        y = player_stats['TS%']
        y=pd.array(y)
        y=y.reshape(-1,1)
        x = player_stats[i]
        x=pd.array(x)
        x=x.reshape(-1,1)
        x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .5)
        model = LinearRegression()
        model.fit(x_train, y_train)
        results.append(model.score(x_test, y_test))
    print(i)
    print('average score from 1000 trials: ' + str(sum(results)/len(results)))
    print()
```
